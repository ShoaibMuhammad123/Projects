## HOW MANY MATCHES HAVE BEEN PLAYED IN EACH CITY

```python
data=ipl.city.value_counts()
print(data.head())

cities=ipl.city.value_counts().keys()
no_of_matches = ipl.city.value_counts()

# plt.title('HOW MANY MATCHES HAVE BEEN PLAYE')
sb.set_style('darkgrid')
sb.relplot(y=cities,x=no_of_matches,hue=data.keys(),data=data)
plt.xlabel('no of matches ',c='b')
plt.ylabel('Cities Name',c='b')
# plt.savefig('no of matches played.jpeg')
plt.show()
```

##  HOW MANY MATCHES HAVE BEEN PLAYED IN EACH SEASON

```python
# How many matches have been played in each season
matches=ipl['season'].value_counts()
print(matches)

sb.set_style('darkgrid')
plt.bar(ipl['season'].value_counts().keys(),ipl['season'].value_counts(),color=['orange','y','r','b','black','c','brown','teal'],width=0.6)

plt.title("IPL SEASONS AND THEIR MATCHES",c='black',fontdict={'fontname':'forte','fontsize':18})

plt.xlabel('IPL SEASONS',c='black',fontdict={'fontname':'forte','fontsize':14})
plt.ylabel('NO OF MATCHES',c='black',fontdict={'fontname':'forte','fontsize':14})
plt.xticks([2008,2009,2010,2011,2012,2013,2014,2015,2016])
# plt.savefig('season and matches.png')
plt.show()
```

### Toss play important role or not

```python

win_by_choosing_field = (ipl[(ipl['toss_decision']=='field')&(ipl[ 'win_by_wickets']!=0)])['id'].count()
print("Number of matches win by choosing field :",win_by_choosing_field)

loss_by_choosing_field= ipl[(ipl['toss_decision']=='field')&ipl['win_by_runs']!=0]['id'].count()
print("Number of matches lossed by choosing field :",loss_by_choosing_field)
```

```python

win_by_choosing_bat = ipl[(ipl['toss_decision']=='bat')&(ipl[ 'win_by_runs']!=0)].shape[0]
print("Number of matches win by choosing field :",win_by_choosing_bat)

loss_by_choosing_bat = ipl[(ipl['toss_decision']=='bat')&ipl['win_by_wickets']!=0].shape[0]
print("Number of matches lossed by choosing bat :",loss_by_choosing_bat)
```

### Choosing field win and loss rate

```python
plt.pie([win_by_choosing_field,loss_by_choosing_field],colors=['r','y'],autopct="%.1f%%",labels=['win_by_choosing_fields','loss_by_choosing_field'])
plt.savefig('loss winning toss')
plt.title('Choosing field win and loss rate')
plt.show()



```

#### Choosing bat win and loss rate

```python
plt.pie([win_by_choosing_bat,loss_by_choosing_bat],colors=['r','y'],autopct="%.1f%%",labels=['win_by_choosing_bat','loss_by_choosing_bat'])
plt.savefig('loss winning toss')
plt.title('Choosing bat win and loss rate')
plt.show() 
```

### Win and Loss Rate by wining toss

```python
# percentage of win and loss after winning toss
plt.pie([win_by_choosing_bat,win_by_choosing_field,loss_by_choosing_field,loss_by_choosing_bat],colors=['r','y','c','b'],
        autopct="%.1f%%",labels=['win by choosing bat','win by choosing field','loss by choosing field','loss by choosing bat'])
plt.title('Rate of wining or loosing  after wining the toss ',fontdict={'fontname':'Cooper Black','fontsize':18},c='y')
plt.savefig('win loss rate')

plt.show()
```

#### Combinly win and loss rate 

```python
win = win_by_choosing_field + win_by_choosing_bat
loss =loss_by_choosing_field +loss_by_choosing_bat


plt.pie([win,loss],colors=['r','c'],autopct="%.1f%%",labels=['wins','loss'])
plt.title('Win and loss Rate by wining toss')
plt.savefig('loss winning toss')
plt.show() 
```
